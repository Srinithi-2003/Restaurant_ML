 * Target Encoding

Reason for using target encoding
. Perserves Information where by transforming into meaningful insights (Numerical Values)
. It maintain the nature of categorical values though transformed as numeric values
. Encoding with mean of the target variable captures the relationship between categorical and target variable by proving a correlation .
. Reduces the Dimensionality

```python
def target_encode(Restaurant,col,target):
    target_map=Restaurant.groupby(col)[target].mean()
    Restaurant[col]=Restaurant[col].map(target_map)
for col in ['City','Address','Locality','Locality Verbose','Cuisines']:
    target_encode(Restaurant,col,'Aggregate rating')
```

```python
X=Restaurant.drop('Aggregate rating',axis=1)
Y=Restaurant['Aggregate rating']
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

Random Forest

Since its a Ensemble learning it builds mutiple decision tree during the training process and improve the genralisation ad also reduce the overfitting . By combining the decison tree random forest improves the robust and provide accurate prediction comapred to induvidual decision tree .

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder
data = pd.read_csv('Restaurant_Analysis.csv')
data = data.drop(columns=[ 'Restaurant Name', 'Currency', 'Rating color', 'Rating text', 'Switch to order menu'])
label_encoder = LabelEncoder()
for col in data.select_dtypes(include=['object']):
    data[col] = label_encoder.fit_transform(data[col])
X = data.drop('Aggregate rating', axis=1)
y = data['Aggregate rating']
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X, y)
feature_importances = rf_model.feature_importances_
print("Feature Importances:")
for feature, importance in zip(X.columns, feature_importances):
    print(f"{feature}: {importance}")
max_importance_idx = feature_importances.argmax()
max_importance = feature_importances[max_importance_idx]
max_feature = X.columns[max_importance_idx]

print("\nMaximum Feature Importance:")
print("Feature:", max_feature)
print("Importance:", max_importance)

```

Using the feature importance score of each column its found that Votes column is the more influential one

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
from sklearn.preprocessing import LabelEncoder
```

Cuisines , City , Price range is been taken the X varaible as its classification of cuisine it could be based on the city and the price for each cuisine

```python
import geopandas as gpd
from shapely.geometry import Point
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

geometry = [Point(lon, lat) for lon, lat in zip(Restaurant['Longitude'], Restaurant['Latitude'])]
gdf = gpd.GeoDataFrame(Restaurant, geometry=geometry, crs='epsg:4326')

fig, ax = plt.subplots(figsize=(10, 10))
World.plot(ax=ax, color='lightgrey')
gdf.plot(ax=ax, marker='o', color='red', markersize=5, label='Restaurant')
plt.legend()


kmeans = KMeans(n_clusters=5)
gdf['cluster'] = kmeans.fit_predict(gdf[['Longitude', 'Latitude']])
centers = kmeans.cluster_centers_
ax.scatter(centers[:, 0], centers[:, 1], marker='x', color='blue', s=100, label='Cluster Centers')
plt.legend()

for i, center in enumerate(centers):
    ax.text(center[0], center[1], f'Cluster {i}', fontsize=10)
plt.show()
```
